## Introduction to pyplot
### Gerando visualizações com pyplot

```python
# importando a biblioteca

import matplotlib.pyplot as plt
```

```python
# plotando um gráfico básico
plt.plot([1, 2, 3, 4, 5])
plt.ylabel('números')
plt.show()
# o gráfico plotado tem y variando de 1 a 5 pois a função plot assume que 
# se há apenas uma sequência de números inseridos nos parenteses, então estes são os valores de y
# a função presume que a sequência de X será também em ordem, tal qual Y, mas
# como Python começa a contar a partir de 0 e não de 1, os valores de X ficam: [0,1,2,3,4]
```

```python
# plotando um gráfico com valores de X e Y fornecidos direto na função

plt.plot([10, 20, 30, 40], [1, 2.4, 38, 16])
# com esta "notação" de inserir duas listas [x1,x2,x3,x4] e [y1,y2,y3,y4] é possível plotar função com valores desejados de x e y
```

```python
# Formatando o gráfico desejado

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
# o terceiro argumento da função, neste caso 'ro', é do tipo string e serve para definir qual a cor e formato da plotagem dos dados
# neste caso, o padrão da função plot é 'b-', que significa b=blue e -=line,
# no caso de 'ro', isto significa: r=red e o=círculo, o que nos gera um gráfico de dispersão de pontos com cor vermelha
plt.axis((0, 6, 0, 20)) #define de quanto a quanto varia o eixo x e o eixo y = plt.axis(x0, x1, y0, y1)
plt.show()
```

```python
#Criando uma série e plotando um gráfico com diferentes formas

# importando numpy
import numpy as np

# Criando uma série temporal que varia de 0 a 5 segundos, em intervalos de 2 milisegundos (0.2 segundos)
t = np.arange(0., 5., 0.2)

# Plotar gráfico tracejado vermelho, quadrados azuis e triangulos verdes
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
